```python
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values('date')
```

```python
goals = {
    'recall': 0.8,
    'precision_val': 0.3,
    'roc_auc': 0.8
}

metrics_to_plot = ['recall', 'precision_val', 'roc_auc']
titles = ['Recall Over Time', 'Precision Over Time', 'ROC AUC Over Time']
colors = ['#1f77b4', '#ff7f0e', '#2ca02c'] 

fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10, 12), sharex=True)
fig.suptitle('Model Performance Metrics vs. Goals Over Time', fontsize=16, y=0.95) 

for i, metric in enumerate(metrics_to_plot):
    ax = axes[i]
    goal = goals[metric]
    color = colors[i]

    ax.plot(df['date'], df[metric], marker='o', linestyle='-', color=color, label=f'{metric.replace("_val", "").capitalize()} Value')

    ax.axhline(goal, color='red', linestyle='--', linewidth=1.5, label=f'Goal: {goal:.1f}')

    ax.fill_between(df['date'], df[metric], goal, where=(df[metric] >= goal),
                    color='green', alpha=0.2, interpolate=True, label='Above Goal')

    ax.fill_between(df['date'], df[metric], goal, where=(df[metric] < goal),
                    color='red', alpha=0.2, interpolate=True, label='Below Goal')

    ax.set_title(titles[i])
    ax.set_ylabel('Score')
    ax.legend()
    ax.grid(True, linestyle=':', alpha=0.7)

    ax.xaxis.set_major_locator(mdates.AutoDateLocator()) 
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

axes[-1].set_xlabel('Date')

plt.xticks(rotation=45, ha='right')

plt.tight_layout(rect=[0, 0.03, 1, 0.93])

plt.show()
```

```python
df_description = df[['recall', 'precision_val', 'roc_auc']].describe()

metrics_to_plot = ['recall', 'precision_val', 'roc_auc']
goals = {
    'recall': 0.8,
    'precision_val': 0.3,
    'roc_auc': 0.8
}

stats_data = {}
stat_rows = ['min', '25%', '50%', '75%', 'max', 'mean']
for metric in metrics_to_plot:
    metric_stats = df_description.loc[stat_rows, metric].to_dict()
    metric_stats['goal'] = goals[metric]
    stats_data[metric] = metric_stats

stat_order = ['min', '25%', '50%', '75%', 'max', 'mean']

titles = ['Recall Distribution Summary', 'Precision Distribution Summary', 'ROC AUC Distribution Summary']
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 6)) 
fig.suptitle('Summary Statistics vs. Goals for Key Metrics (from describe())', fontsize=16, y=1.02) 

for i, metric in enumerate(metrics_to_plot):
    ax = axes[i]
    metric_data = stats_data[metric]
    goal = metric_data['goal']
    color = colors[i]
    
    values = [metric_data[stat] for stat in stat_order]

    bars = ax.bar(stat_order, values, color=color, alpha=0.7)

    ax.axhline(goal, color='red', linestyle='--', linewidth=1.5, label=f'Goal: {goal:.1f}')

    ax.bar_label(bars, fmt='%.3f', padding=3, fontsize=9)

    ax.set_title(titles[i])
    ax.set_ylabel('Score')
    ax.set_xlabel('Statistic')
    ax.legend()
    ax.grid(True, axis='y', linestyle=':', alpha=0.7) 

    plot_max_value = max(values + [goal])
    ax.set_ylim(bottom=0, top=plot_max_value * 1.15) 

plt.tight_layout(rect=[0, 0, 1, 0.95]) 
plt.show()
```
